```python
#1clean rate column
# Replace 'NEW' and '-' with NaN (handle as missing)
df['rate'] = df['rate'].replace(['NEW', '-','nan'], np.nan)

# Extract only numeric values like '4.1' from '4.1/5' (if present)
df['rate'] = df['rate'].astype(str).str.strip().str.replace('/5', '', regex=False)

# Convert the result to float (now it's safe)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Fill remaining NaN values with the column's mean
df['rate'] = df['rate'].fillna(df['rate'].mean())

# Optional: Round to 1 decimal place
df['rate'] = df['rate'].round(1)
```

```python
print(df['rate'].unique()[:10])    # See some sample cleaned rates
```

```python
df.isnull().sum()   
```

```python
#2. Clean and Fill 'approx_cost(for two people)'
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype(str).str.replace(',','',regex=True)
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype(float)
df['approx_cost(for two people)']=df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].mean())    
```

```python
#3. Fill 'rest_type', 'location', 'phone', 'cuisines', 'dish_liked'
df['rest_type']= df['rest_type'].fillna('Not Given' )
df['location']= df['location'].fillna('Not Available')
df['phone']= df['phone'].fillna('unkonwn')
df['cuisines']= df['cuisines'].fillna('No info')
df['dish_liked']= df['dish_liked'].fillna('others')
```

```python
df.isnull().sum()   
```

```python
#rename a col
df.rename(columns={'listed_in(city)': 'city'}, inplace=True)
df.rename(columns={'approx_cost(for two people)':'cost_for_two_people'}, inplace=True)
df.rename(columns={'listed_in(type)': 'type'}, inplace=True)
```

```python
#Step 6:Univariate Analysis
# Histogram of rating
plt.figure(figsize=(10,5))
sns.histplot(df['rate'].dropna().astype(float), bins=20, kde=True,color='purple')  
plt.title('Distribution of Restaurant Ratings')    
plt.xlabel('Rating')
plt.ylabel('Count')  
plt.show()                                      
```

```python
#step7:Bivariate Analysis
# Rating vs Price
plt.figure(figsize=(8,5))
sns.scatterplot(x='cost_for_two_people', y='rate', data=df,color='green')
plt.title('Rating vs Approx Cost')
plt.show()
```

```python
#Step 8:Correlation Analysis
# Correlation matrix
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='viridis')#viridis,coolwarn,tab20c,tab10
plt.title('Correlation Matrix')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Step 1: Join all cuisine names into one large string
text = ' '.join(df['cuisines'].dropna().astype(str))

# Step 2: Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Set1').generate(text)

# Step 3: Display it
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Popular Cuisines WordCloud')
plt.show()

#explanation:
#Top Cuisines (Most Frequent):
#The largest and most prominent words â€” North Indian, Indian, Chinese, South Indian, Fast Food, Biryani â€” indicate these are the most frequently mentioned cuisines across restaurants, showing dominant customer preferences.
#Moderately Popular Cuisines:
#Medium-sized terms like Desserts, Ice Cream, Bakery, Mughlai, Continental, Cafe suggest these cuisines are also popular but not as dominant as the top ones.
#Less Common Cuisines:
#Smaller words such as Chinese Momos, Street Food, Italian, American, Finger Food represent niche or less frequently listed cuisines, possibly indicating unique offerings or limited presence.
```
